```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 20          # Total number of trials (calls)
p = 0.20        # Probability of success on a single trial
k_success = 4   # Desired number of successful calls

# Calculate the probability using binomial distribution
probability_of_success = binom.pmf(k_success, n, p)

# Print the probability with higher precision
print(f"Probability of exactly {k_success} successful calls: {probability_of_success:.5f}")

# Visualizing the distribution
k_values = np.arange(0, n + 1)
probabilities = binom.pmf(k_values, n, p)

# Create the plot
plt.figure(figsize=(8, 6))
plt.bar(k_values, probabilities, color='skyblue')
plt.title('Binomial Distribution: Number of Successful Sales Calls')
plt.xlabel('Number of Successful Calls')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y')

# Highlight the bar for exactly 4 successful calls
plt.bar(k_success, probabilities[k_success], color='orange', label=f'P(X={k_success})', edgecolor='black')

plt.legend()
plt.show()
```
